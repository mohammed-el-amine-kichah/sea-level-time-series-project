```r
data <- read.csv("algiers_weather.csv")
ts_data <- data[, c("time", "temperature_mean")]
```

```r
library(TSA)
ts_temp <- ts(ts_data$temperature_mean, frequency = 365, start = c(2002, 1))
```

```r
ts_data$time <- as.Date(ts_data$time)
```

```r
par(bg = "white", cex.main = 1.5, cex.lab = 1.8, cex.axis = 1.9)
plot(ts_temp, main = "Daily Mean Temperature in Algiers (20022023)", ylab = "Temperature (C)")
```

```r
par(bg = "white")
TSA::acf(ts_temp, main = "ACF of Daily Mean Temperature", lag.max = 365)
```

```r
tseries::adf.test(ts_temp, alternative = "stationary", k = 0)
```

```r

tseries::adf.test(ts_temp, alternative = "stationary", k = trunc((length(ts_temp)-1)^(1/3)) )
```

```r
par(bg = "white")
stats::pacf(ts_temp, lag.max = 365, main = "PACF of Daily Mean Temperature in Algiers", ylab = "PACF")
```

```r
class(ts_temp)
frequency(ts_temp)  # Should return 365
```

```r
par(bg = "white")
stats::pacf(ts_temp, lag.max = 730, plot.type = "histogram", main = "PACF of Daily Mean Temperature in Algiers", ylab = "PACF")
```
